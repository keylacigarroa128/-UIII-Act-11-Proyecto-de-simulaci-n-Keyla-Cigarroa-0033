### Estilistas ✂️💇‍♀️

```python
# --- PROGRAMA DE SIMULACIÓN DE IA: GLAMOUR DATA-ANALYTICS ---

# El diccionario principal empieza completamente vacío
diccionario_estilistas = {}


def registrar_estilista():
    print("\n--- 📝 REGISTRO DE NUEVO ESTILISTA ---")

    # Entrada del nombre que servirá como la LLAVE (Key) principal
    nombre = input("Nombre del estilista: ")

    # Entrada de los valores (Values)
    especialidad = input("Especialidad en color (ej. Balayage): ")
    salon = input("Salón de belleza al que pertenece: ")
    agenda = input("Agenda semanal (ej. Lun-Vie 9-18): ")
    tecnicas = input("Técnicas dominadas: ")
    marcas = input("Marcas aliadas: ")
    premios = input("Premios o reconocimientos: ")

    # Guardamos los datos dentro del diccionario principal
    diccionario_estilistas[nombre] = {
        "Especialidad en color": especialidad,
        "Salón de belleza": salon,
        "Agenda semanal": agenda,
        "Técnicas dominadas": tecnicas,
        "Marcas aliadas": marcas,
        "Premios": premios,
    }

    print(f"\n✅ ¡IA: Estilista '{nombre}' registrado con éxito!")


def mostrar_estilistas():
    print("\n" + "==================================================")
    print(" 💈 BASE DE DATOS: DIRECTORIO DE ESTILISTAS 💈")
    print("==================================================")

    if len(diccionario_estilistas) == 0:
        print("La base de datos está vacía. ¡Registra a alguien primero!")
    else:
        # Recorremos el diccionario principal
        for nombre, datos in diccionario_estilistas.items():

            # Imprimimos el nombre del estilista como encabezado de su propia tarjeta
            print(f"📌 ESTILISTA: {nombre.upper()}")
            print("-" * 50)

            # Imprimimos cada llave y valor alineados a la izquierda con un ancho fijo de 22 caracteres
            print(f"{'  • Especialidad:':<22} {datos['Especialidad en color']}")
            print(f"{'  • Salón de belleza:':<22} {datos['Salón de belleza']}")
            print(f"{'  • Agenda semanal:':<22} {datos['Agenda semanal']}")
            print(f"{'  • Técnicas:':<22} {datos['Técnicas dominadas']}")
            print(f"{'  • Marcas aliadas:':<22} {datos['Marcas aliadas']}")
            print(f"{'  • Premios:':<22} {datos['Premios']}")
            print("=" * 50) # Línea de cierre para cada estilista

    print("\n")


def simular_ia_recomendacion():
    print("\n--- 🤖 MÓDULO DE IA: ASISTENTE RECOMENDADOR ---")

    if len(diccionario_estilistas) == 0:
        print("IA: No tengo datos en el diccionario para analizar.")
    else:
        print("IA: ¿Qué técnica, marca o especialidad estás buscando hoy?")
        busqueda = input("Introduce una palabra clave (ej. Balayage): ")
        print("\nIA: Procesando diccionarios y buscando coincidencias...")

        encontrado = False

        for nombre, datos in diccionario_estilistas.items():
            if (
                busqueda.lower() in datos["Especialidad en color"].lower()
                or busqueda.lower() in datos["Técnicas dominadas"].lower()
                or busqueda.lower() in datos["Marcas aliadas"].lower()
            ):
                print(f"⭐ ¡RECOMENDACIÓN IA!: Te sugerimos a **{nombre}**")
                print(f"   Salón: {datos['Salón de belleza']}")
                print(f"   Habilidades: {datos['Técnicas dominadas']}")
                print(f"   Agenda disponible: {datos['Agenda semanal']}\n")
                encontrado = True

        if not encontrado:
            print("IA: No encontré estilistas con esa combinación de habilidades.")


# --- MENÚ INTERACTIVO PRINCIPAL (Ciclo While True) ---
while True:
    print("========================================")
    print("      💇 CONSOLA DE INTELIGENCIA ARTIFICIAL 💇")
    print("            'GLAMOUR DATA-ANALYTICS'   ")
    print("========================================")
    print("1. Registrar nuevo estilista")
    print("2. Mostrar directorio de estilistas (Alineado)")
    print("3. Consultar al Asistente de IA (Búsqueda)")
    print("4. Salir del sistema")
    print("========================================")

    opcion = input("Elige una opción (1-4): ")

    if opcion == "1":
        registrar_estilista()
    elif opcion == "2":
        mostrar_estilistas()
    elif opcion == "3":
        simular_ia_recomendacion()
    elif opcion == "4":
        print("\n👋 ¡Gracias por usar la IA de Glamour Data-Analytics! Cerrando sistemas...")
        print("programa realizado por keyla cigarroa nl = 12}}")
        break
    else:
        print("❌ Opción no válida. Por favor, elige un número del 1 al 4.\n")
```
