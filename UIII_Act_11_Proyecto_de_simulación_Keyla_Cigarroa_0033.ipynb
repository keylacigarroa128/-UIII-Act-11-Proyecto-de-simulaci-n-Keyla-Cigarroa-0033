### Estilistas ✂️💇‍♀️

```python
# --- PROGRAMA DE SIMULACIÓN DE IA PARA SALÓN DE BELLEZA ---

# Al inicio, la lista que contendrá los diccionarios empieza vacía
lista_estilistas = []


def registrar_estilista():
    print("\n--- 📝 REGISTRO DE NUEVO ESTILISTA ---")

    # Creamos el diccionario vacío para el nuevo estilista
    estilista = {}

    # Entrada de datos (Key, Values)
    nombre = input("Nombre del estilista: ")
    especialidad = input("Especialidad en color (ej. Balayage): ")
    salon = input("Salón de belleza al que pertenece: ")
    agenda = input("Agenda semanal (ej. Lun-Vie 9-18): ")
    tecnicas = input("Técnicas dominadas: ")
    marcas = input("Marcas aliadas: ")
    premios = input("Premios o reconocimientos: ")

    # Asignamos los valores a sus respectivas llaves (Keys)
    estilista["Nombre"] = nombre
    estilista["Especialidad en color"] = especialidad
    estilista["Salón de belleza"] = salon
    estilista["Agenda semanal"] = agenda
    estilista["Técnicas dominadas"] = tecnicas
    estilista["Marcas aliadas"] = marcas
    estilista["Premios"] = premios

    # Agregamos el diccionario a la lista
    lista_estilistas.append(estilista)
    print(f"\n✅ ¡IA: Estilista '{nombre}' registrado con éxito!")


def mostrar_estilistas():
    print("\n" + "=" * 135)
    print(" 💈 DIRECTORY GENERAL DE ESTILISTAS REGISTRADOS 💈")
    print("=" * 135)

    if len(lista_estilistas) == 0:
        print("La base de datos está vacía. ¡Registra a un estilista primero!")
    else:
        # ENCABEZADO DE LA TABLA: El símbolo '<' alinea a la izquierda y el número define el ancho de la columna
        encabezado = f"{'Nombre':<15} | {'Especialidad':<20} | {'Salón':<20} | {'Agenda Semanal':<18} | {'Técnicas':<20} | {'Marcas':<18} | {'Premios':<15}"
        print(encabezado)
        print("-" * 135)  # Línea divisoria de la tabla

        # Ciclo FOR para recorrer la lista de diccionarios e imprimir las filas alineadas
        for estilista in lista_estilistas:
            fila = (
                f"{estilista['Nombre']:<15} | "
                f"{estilista['Especialidad en color']:<20} | "
                f"{estilista['Salón de belleza']:<20} | "
                f"{estilista['Agenda semanal']:<18} | "
                f"{estilista['Técnicas dominadas']:<20} | "
                f"{estilista['Marcas aliadas']:<18} | "
                f"{estilista['Premios']:<15}"
            )
            print(fila)

    print("=" * 135 + "\n")


def simular_ia_recomendacion():
    print("\n--- 🤖 MÓDULO DE IA: ASISTENTE RECOMENDADOR ---")

    if len(lista_estilistas) == 0:
        print("IA: No tengo datos suficientes. Registra estilistas primero.")
    else:
        print("IA: ¿Qué técnica, marca o especialidad estás buscando hoy?")
        busqueda = input("Introduce una palabra clave (ej. Balayage): ")
        print("\nIA: Procesando datos y buscando coincidencias...")

        encontrado = False
        for estilista in lista_estilistas:
            # Evaluamos condiciones usando IF y operadores lógicos
            if (
                busqueda.lower() in estilista["Especialidad en color"].lower()
                or busqueda.lower() in estilista["Técnicas dominadas"].lower()
                or busqueda.lower() in estilista["Marcas aliadas"].lower()
            ):
                print(
                    f"⭐ ¡RECOMENDACIÓN IA!: Te sugerimos a **{estilista['Nombre']}** en el salón '{estilista['Salón de belleza']}'."
                )
                print(f"   Habilidades: {estilista['Técnicas dominadas']}")
                print(f"   Agenda disponible: {estilista['Agenda semanal']}\n")
                encontrado = True

        if not encontrado:
            print("IA: No encontré estilistas con esa combinación de habilidades.")


# --- MENÚ INTERACTIVO PRINCIPAL (Ciclo While True) ---
while True:
    print("========================================")
    print("      💇 CONSOLA DE INTELIGENCIA ARTIFICIAL 💇")
    print("            'GLAMOUR DATA-ANALYTICS'   ")
    print("========================================")
    print("1. Registrar nuevo estilista")
    print("2. Mostrar tabla de estilistas (Alineada)")
    print("3. Consultar al Asistente de IA")
    print("4. Salir del sistema")
    print("========================================")

    opcion = input("Elige una opción (1-4): ")

    # Estructura de control IF, ELIF, ELSE
    if opcion == "1":
        registrar_estilista()
    elif opcion == "2":
        mostrar_estilistas()
    elif opcion == "3":
        simular_ia_reconendacion() if "simular_ia_reconendacion" in globals() else simular_ia_recomendacion()
    elif opcion == "4":
        print("\n👋 ¡Gracias por usar la IA de Glamour Data-Analytics! Cerrando sistemas...")
        print("Programa realizado por Keyla Cigarroa NL = 12")
        break
    else:
        print("❌ Opción no válida. Por favor, elige un número del 1 al 4.\n")
```
